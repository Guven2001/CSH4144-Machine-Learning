```python
#print matrix
m=np.array([[1,3,2],[2,5,4],[4,6,2]])
print(m)
```

```python
#Transpose
print('Matrix Transpose: \n',m.transpose(),'\n')
```

```python
#Determinant
print('Matrix Determinant: \n',np.linalg.det(m),'\n')
```

```python
#Inverse
m_inv=np.linalg.inv(m)
print('Matrix Inverse: \n',m_inv,'\n')
```

```python
iden_m = np.dot(m,m_inv)
iden_m = np.round(np.abs(iden_m),0)
print("Product of Matrix and its inverse: \n",iden_m)
```

```python
#Eigen Values and Vectors
eigen_values,eigen_vectors = np.linalg.eig(m)
print('Eigen values',eigen_values,'\n')
print('Eigen vectors \n',eigen_vectors,'\n')
```

```python
U,S,VT=np.linalg.svd(m)
print('U : \n',U,'\n')
print('S : \n',S,'\n')
print('VT : \n',VT,'\n')
```

```python
#get data
import scipy as sp
import numpy as np

nums=np.random.randint(1,25, size=(1,15))[0]
print('Data : ',nums)
```

```python
# Get Descriptive Statistics

print('Mean: ',np.mean(nums),'\n')
# or print('Mean: ',sp.mean(nums),'\n')
print('Median: ',np.median(nums),'\n')
# or print('Median: ',sp.median(nums),'\n')

print('Mode: ',sp.stats.mode(nums),'\n')
# return unbiased skew 

print('Skew: ',sp.stats.skew(nums),'\n')
print('Kurtosis: ',sp.stats.kurtosis(nums),'\n')

print('Standard Deviation: ',sp.std(nums),'\n')

print('Variance: ',np.var(nums),'\n')


```
